### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [25, 35, 30, 10]
scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot line chart in the first subplot
axs[0, 0].plot(x, y, label='Sin(x)', color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].legend()

# Plot bar chart in the second subplot
axs[0, 1].bar(categories, values, color='green')
axs[0, 1].set_title('Bar Chart')

# Plot scatter plot in the third subplot
axs[1, 0].scatter(scatter_x, scatter_y, color='red')
axs[1, 0].set_title('Scatter Plot')

# Plot pie chart in the fourth subplot
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```
